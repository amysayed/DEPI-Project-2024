```python
colors=['darkblue', 'darkgreen', 'darkorange', 'red',"orchid"]
Number_of_Children=pd.DataFrame(df["Number of Children"].value_counts())
Number_of_Children.columns = ['Total Count']  # change column count into total count 
Number_of_Children=Number_of_Children.reset_index()
plt.figure(figsize=(18,6))
sns.barplot(data=Number_of_Children,x="Number of Children",y="Total Count",palette=colors)
plt.title("Number of Children effect On Depression")
plt.ylabel('Depressed People',fontsize=14,weight='bold')
plt.show()
```

```python
Marital_Status=pd.DataFrame(df["Marital Status"].value_counts())
Marital_Status.columns = ['Total Count']  # change column count into total count 
Marital_Status=Marital_Status.reset_index()
Marital_Status
```

```python
colors=['darkblue', 'darkgreen', 'darkorange', 'red',"orchid"]
plt.figure(figsize=(18,6))
sns.barplot(data=Marital_Status,x="Marital Status",y="Total Count",hue="Marital Status", palette=colors)
plt.title("Marital_Status effect On Depression")
plt.ylabel('Depressed People',fontsize=14,weight='bold')
plt.xticks(rotation=45)
plt.show()
```

#### Is marital status associated with depression risk?

```python
Education_Level=pd.DataFrame(df["Education Level"].value_counts())
Education_Level.columns = ['Total Count']  # change column count into total count 
Education_Level=Education_Level.reset_index()
plt.figure(figsize=(18,6))
sns.barplot(data=Education_Level,x="Education Level",y="Total Count",palette=colors)
plt.title("Education_Level effect On Depression")
plt.ylabel('Depressed People',fontsize=14,weight='bold')
plt.xticks(rotation=45)
plt.show()
```
